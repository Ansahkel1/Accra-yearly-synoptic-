```python
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
#import seaborn as sns
#from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
#import cartopy.feature as cfeeature

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
Accra_max = xr.open_dataset("Accra_max.nc")
Accra_max
```

```python
Accra_min = xr.open_dataset("Accra_min.nc")
Accra_min
```

```python
yearly_avg_max_temp_1 = Accra_max['mx2t'].groupby('time.year').mean(dim='time')
yearly_avg_max_temp_1
```

```python
yearly_avg_min_temp_2 = Accra_min['mn2t'].groupby('time.year').mean(dim='time')
yearly_avg_min_temp_2
```

```python
# Plotting the yearly climatological temperature averages for maximum temperature
plt.figure(figsize=(12, 6))
yearly_avg_max_temp_1.plot(marker='o', color='red')
plt.title('Yearly Average Maximum Temperature in Accra from 1979 to 2020\n')
plt.xlabel('Year')
plt.ylabel('Temperature (K)')
plt.grid(True)
plt.xticks(yearly_avg_max_temp_1.year, rotation=95)
plt.legend(['Max Temperature'])
plt.show()
```

```python
# Plotting the yearly climatological temperature averages for minimum temperature
plt.figure(figsize=(10, 5))
yearly_avg_min_temp_2.plot(marker='o', color='blue')
plt.title('Yearly Average Minimum Temperature in Accra from 1979 to 2020\n')
plt.xlabel('Year')
plt.ylabel('Temperature (K)')
plt.grid(True)
plt.xticks(yearly_avg_min_temp_2.year, rotation=95)
plt.legend(['Min Temperature'])
plt.show()
```
